```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Set random seed for reproducibility
np.random.seed(0)

# Generate non-normally distributed data (exponential distribution)
df_filled = pd.DataFrame({
    'quad': np.random.exponential(scale=2.0, size=1000)  # Exponential distribution
})

# Initialize and fit StandardScaler
SS_scaler = StandardScaler()
df_filled['SS_quad'] = SS_scaler.fit_transform(df_filled[['quad']])

# Plot original and standardized data
plt.figure(figsize=(14, 7))

# Original Data
plt.subplot(1, 2, 1)
sns.histplot(df_filled['quad'], kde=True, color='blue', bins=30)
plt.title('Original Data (Exponential Distribution)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Standardized Data
plt.subplot(1, 2, 2)
sns.histplot(df_filled['SS_quad'], kde=True, color='red', bins=30)
plt.title('Standardized Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
