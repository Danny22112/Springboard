<h1> Number Of Missing Values By Column

```python
churn_rate.isna().sum()
```

```python
missing = pd.concat([churn_rate.isnull().sum(), 100 * churn_rate.isnull().mean()], axis=1)
missing.columns=['Count', '%']
missing.sort_values(by='Count', ascending=False)
```

```python
# Each missing value by row
```

```python
churn_rate_Total_Charges = churn_rate[churn_rate.isna().any(axis=1)]
print(churn_rate_Total_Charges)
```

<h1> Replace Total Charges with statistical mean

```python
total_charges_mean = churn_rate['Total Charges'].mean()
```

```python
churn_rate_new = churn_rate.fillna({'Total Charges': total_charges_mean})
```

```python
churn_rate_new.isna().sum()
```
